**Data Preprocessing**

Feature Engineering

```python
import random
import time
```

```python
df
```

```python
#getting rid of columns that we found faulty
df=df.drop(columns=['months_employed'])
```

```python
#we have 2 columns personal_account_m and personal_account_y
#combining both columns 
df['personal_account_months'] = (df.personal_account_m+(df.personal_account_y*12))
```

```python
df[['personal_account_months','personal_account_m','personal_account_y']].head()
```

```python
df=df.drop(columns=['personal_account_m','personal_account_y'])
```

One Hot Encoding

```python
df = pd.get_dummies(df)
```

```python
df.columns
```

```python
#we need to remove one of the newly created columns to avoid the DUMMY VARIABLE TRAP
df=df.drop(columns = ['pay_schedule_semi-monthly'])
```

Removing Extra Columns

```python
response = df['e_signed']
user_identifier = df['entry_id']
df=df.drop(columns=['e_signed','entry_id'])
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df,response,test_size=0.2, random_state=0)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = pd.DataFrame(sc.fit_transform(X_train))
```

```python
X_test_scaled = pd.DataFrame(sc.transform(X_test))
```

```python
X_train_scaled.columns = X_train.columns.values
X_test_scaled.columns = X_test.columns.values
X_train_scaled.index = X_train.index.values
X_test_scaled.index = X_test.index.values
```

```python
X_train = X_train_scaled
X_test = X_test_scaled
```

**Model Building**

Comparing Models

1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0,penalty='l1',solver='liblinear')  #lasso penalty
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
acc=accuracy_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
```

```python
results = pd.DataFrame([['Logistic Regression (Lasso)',acc,prec,rec,f1]],columns=['Model','Accuracy','Precision','Recall','F1'])
results
```

2. SVM (Linear)

```python
from sklearn.svm import SVC
classifier = SVC(random_state=0, kernel='linear') 
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
acc=accuracy_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
model_results = pd.DataFrame([['SVM (Linear)',acc,prec,rec,f1]],columns=['Model','Accuracy','Precision','Recall','F1'])
results = results.append(model_results, ignore_index=True)
results
```

3. SVM (RBF)

```python
from sklearn.svm import SVC
classifier = SVC(random_state=0, kernel='rbf') 
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
acc=accuracy_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
model_results = pd.DataFrame([['SVM (RBF)',acc,prec,rec,f1]],columns=['Model','Accuracy','Precision','Recall','F1'])
results = results.append(model_results, ignore_index=True)
results
```

4. Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(random_state=0, n_estimators=100, criterion='entropy') 
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
acc=accuracy_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
model_results = pd.DataFrame([['Random Forest',acc,prec,rec,f1]],columns=['Model','Accuracy','Precision','Recall','F1'])
results = results.append(model_results, ignore_index=True)
results
```

Saving the Results

```python
results.to_csv('model_results.csv', index=False)
```

K-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)
print("Accuracy: %0.2f"%(accuracies.mean()))
```

**Improving the model**

Grid Search

Round 1: Entropy

```python
from sklearn.model_selection import GridSearchCV
parameters = {'max_depth':[3,None],'max_features':[1,5,10],'min_samples_split':[2,5,10],'min_samples_leaf':[1,5,10],"bootstrap":[True, False],'criterion':['entropy']}

grid_search = GridSearchCV(estimator=classifier, param_grid = parameters,scoring='accuracy',cv=10,n_jobs=-1)   #n_jobs takes a toll on your computer if changed to -1
```

```python
t0=time.time()
grid_search = grid_search.fit(X_train,y_train)
t1=time.time()
print('%0.2f deconds taken'%(t1-t0))
```

```python
rf_best_accuracy = grid_search.best_score_
rf_best_accuracy
```

```python
rf_best_parameters = grid_search.best_params_
rf_best_parameters
```

Round 2: Testing new parameters on Test Set

```python
y_pred = grid_search.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
acc=accuracy_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
model_results = pd.DataFrame([['Random Forest (Grid Search: Entropy)',acc,prec,rec,f1]],columns=['Model','Accuracy','Precision','Recall','F1'])
results = results.append(model_results, ignore_index=True)
results
```
