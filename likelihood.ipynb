Histograms

```python
df_temp = df.drop(columns=['entry_id','pay_schedule','e_signed'])
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Histograms of Numerical Columns',fontsize=20)
for i in range(1,df_temp.shape[1]+1):
    plt.subplot(6,3,i) 
    f = plt.gca()
    f.set_title(df_temp.columns.values[i-1])

    vals = np.size(df_temp.iloc[:,i-1].unique())
    if vals>=100:
        vals=100

    plt.hist(df_temp.iloc[:,i-1],bins=vals,color='#3F5D7D')

plt.tight_layout(rect = [0,0.03,1,0.95])
```

```python
#we see that there are several people in the months_employed feature that have 0 months employed. This is very unusual and leads to the fact that this feature can be inaccurate.
```

Correlation Plot (with response variable)

```python
c = ['red', 'yellow', 'black', 'blue', 'orange','green','violet']
df_temp.corrwith(df['e_signed']).plot(kind='bar',figsize=(20,10),title='Correlation with E Signed Column',fontsize=15,rot=45,grid=True,color=c)
```

```python
#this correlation is linear in nature so this may change when we run models not linear in nature like random forest, SVM, etc.
```

Correlation Matrix

```python
sns.set(style='white',font_scale = 2)     #background
corr = df_temp.corr()    #actual correlation

f,ax = plt.subplots(figsize=(18,15))    #set up matplotlib figure
f.suptitle("Correlation Matrix",fontsize=48)

cmap = sns.diverging_palette(220,10,as_cmap=True)     #set up a custom diverging colormap

sns.heatmap(corr,cmap=cmap,vmax=0.3,center=0,square=True,linewidths=0.5,cbar_kws={"shrink":0.5})
```
